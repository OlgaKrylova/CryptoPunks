```python
import pandas as pd
import os
```

```python
# Reading json file with dataset description
df=pd.read_json('original_dataset/txn_history-2021-10-07.jsonl', lines=True)
```

```python
df.head()
```

```python
# Drop all the unnecessary columns
df = df[['punk_id', 'type']]
```

```python
# Transform 'type' column notes to 'string' type
df['type1'] = df['type'].apply(lambda x: str(x))
```

```python
df = df.drop('type', axis=1)
```

```python
df = df.drop_duplicates()
```

```python
# Number of pictures
df['type1'].value_counts()
```

```python
# Transform dataframe to a pictures type dictionary 
d1 = df.set_index('punk_id').to_dict()
typedict = d1['type1']
typedict
```

```python
path_from = 'original_dataset/imgs/imgs/'
path_to = 'sorted_dataset/imgs/'
```

```python
# Remove images
for i, f in enumerate(sorted(os.listdir(path_from))[1:]):
    file = path_from + '/' + f
    print(file)
    if typedict[int(f.strip('.png'))] == "['Female']":
        os.replace(file, path_to + 'female/'+f)
    elif typedict[int(f.strip('.png'))] == "['Male']":
        os.replace(file, path_to + 'male/'+f)
    elif typedict[int(f.strip('.png'))] == "['Ape']":
        os.replace(file, path_to + 'ape/'+f)
    elif typedict[int(f.strip('.png'))] == "['Zombie']":
        os.replace(file, path_to + 'zombie/'+f)
    elif typedict[int(f.strip('.png'))] == "['Alien']":
        os.replace(file, path_to + 'alien/'+f)
```

```python
# Check created folders
sorted(os.listdir(path_to+'female'))
```
